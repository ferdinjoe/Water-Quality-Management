```python
def trim_data(data, batch_size):
  """
  Trims out the extra data that does not fit into batches
  EX:
  dim, batch_size = 192, 50
  to
  dim, batch_size = 150, 50

  PARAMETERS
  ----------
  data: ndarray
    The data that you want to trim
  batch_size: int
    The batch size
  """
  extra = data.shape[0] % batch_size
  if (extra == 0):
    return data
  return data[:-extra]
```

```python
from typing import *
from sklearn.metrics import mean_squared_error
from keras.models import Sequential

import numpy as np
import tensorflow as tf
import sklearn.metrics as metrics

class StandardTrainer:
    def __init__(
        self,
        model: Sequential,
        X_train: np.ndarray,
        X_test: np.ndarray,
        y_train: np.ndarray,
        y_test: np.ndarray,
    ):
        """
        Trains the model and logs the loss, accuracy, MAE, MSE, and RMSE

        PARAMETERS
        ==========
        model: Sequential
          A Keras Sequential Model
        X_train: ndarray
          A numpy array containing the X training data
        X_test: ndarray
          A numpy array containing the X testing data
        y_train: ndarray
          A numpy array containing the y training data
        y_test: ndarray
          A numpy array containing the y testing data
        """
        self.model      = model
        self.X_train    = X_train
        self.X_test     = X_test
        self.y_train    = y_train
        self.y_test     = y_test
        self.checkpoint = None
        self.callbacks  = []

        self.checkpoint_output_file = "model_checkpoint.hdf5"
  
    def set_checkpoint(self, checkpoint: tf.keras.callbacks.ModelCheckpoint):
        """
        Sets a custom checkpoint configuration for the model.
        The default checkpoint monitors the loss, saves the weights only,
        saves the best result only, and uses the min mode.

        PARAMETERS
        ==========
        checkpoint: tf.keras.callbacks.ModelCheckpoint
          The checkpoint object
        """
        self.checkpoint = checkpoint
  
    def initialize_default_checkpoint(self):
        """
        Initializes the default checkout of the model
        """
        self.checkpoint = tf.keras.callbacks.ModelCheckpoint(
            self.checkpoint_output_file,
            monitor='loss',
            save_weights_only=True,
            save_best_only = True,
            mode = 'min'
        );
  
    def set_checkpoint_output_file(self, filename: str):
        """
        Sets the name of the output file

        PARAMETERS
        ==========
        filename: str
          The name of the file
        """
        self.checkpoint_output_file = filename

    def add_callback(self, callback):
        """
        Adds a callback to the model training process

        PARAMETERS
        ==========
        callback: A Keras Callback
          The callback of the model
        """
        self.callbacks.append(callback);
    
    def train(
        self,
        batch_size: int = 100,
        epochs: int     = 100,
        log_every: int  = 100,
        verbose: int    = 0
    ):
        """
        Trains the model

        PARAMETERS
        ==========
        batch_size: int
          DEFAULT = 1
          The batch size
        epochs: int
          DEFAULT = 1000
          The MAXIMUM epochs you want to train the model for
        log_every: int
          DEFAULT = 100
          Determines when to log the model
        verbose: int
          DEFAULT = 0
          Selects the verboseness of the training process
          0: Log only
          1: Shows Progress Bar
          2: One Line Per epoch
        """

        # Initializing the checkpoint and callback functions
        #if self.checkpoint is None:
        #  self.initialize_default_checkpoint();
        if self.checkpoint is not None:
          self.callbacks.append(self.checkpoint)

        # Training the model
        for i in range(0, epochs+1, log_every):
          print("========================================================")
          print(f"Epoch {i}-{i+log_every}:")
          hist = self.model.fit(
              self.X_train,
              self.y_train,
              batch_size = batch_size,
              epochs = log_every,
              callbacks = self.callbacks,
              verbose = verbose
          );

          # Logs the result
          prediction_result = self.model.predict(self.X_test)
          self.mae = metrics.mean_absolute_error(self.y_test, prediction_result)
          self.rms = mean_squared_error(self.y_test, prediction_result, squared=False)
          self.mse = mean_squared_error(self.y_test, prediction_result)
          self.r2 = metrics.r2_score(self.y_test, prediction_result)

          self.loss = hist.history['loss'][-1]
          self.acc  = hist.history['accuracy'][-1]

          print(f"Accuracy: {hist.history['accuracy'][-1]}")
          print(f"Loss: {hist.history['loss'][-1]}")
          print(f"MAE:{self.mae}")
          print(f"MSE:{self.mse}")
          print(f"RMSE:{self.rms}")
          print(f"R-Squared:{self.r2}")
          print("========================================================")
```

```python
def toybox(X_train, X_test, y_train, y_test, epochs, log_every, verbose, output_csv):
  activation_list = ["relu", "sigmoid", "softmax", "softplus", "softsign",
                     "tanh", "selu", "elu"]
  optimizer_list = ["RMSprop", "Adam", "Adadelta", "Adagrad", "Adamax",
                    "Nadam", "Ftrl"] 
  # SGD CANNOT BE USED DUE TO INCOMPATIBLE DATATYPES
  # TRY IT YOURSELF

  output_file = open(output_csv, "w")

  headers = ",".join(['Optimizer','Activation Function','Activation Function 2','Epoch','Training Loss','MAE','MSE','RMSE','R-Squared'])
  headers += "\n"
  output_file.write(headers)

  output_file.close();
  
  for opt in optimizer_list:
    for act in activation_list:
      for act2 in activation_list:
        print(f"Optimizer: {opt}")
        print(f"Activation Function: {act}")
        print(f"Activation Function 2: {act2}")

        # DEFINE THE MODEL HERE
        model = Sequential()
        model.add(LSTM(200, activation=act, input_shape=(n_steps, n_features), return_sequences=True))
        model.add(LSTM(200, activation=act2))
        model.add(Dense(1))
        model.compile(optimizer = opt, loss = "mse", metrics = ['accuracy'])

        # START THE TRAINING PROCESS
        trainer = StandardTrainer (
            model = model,
            X_train = X_train,
            X_test = X_test,
            y_train = y_train,
            y_test = y_test
        )

        trainer.train (
            batch_size = b_size,
            epochs = epochs,
            log_every = log_every,
            verbose = verbose
        )

        output_str = [opt, act, act2, epochs, trainer.loss, trainer.mae, trainer.mse, trainer.rms, trainer.r2]
        output_str = [str(i) for i in output_str]
        output_str = ",".join(output_str)
        output_str += "\n"

        output_file = open(output_csv, 'a')
        output_file.write(output_str)
        output_file.close()

```
